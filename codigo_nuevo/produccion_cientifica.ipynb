```python
import pandas as pd
import os
import numpy as np
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import matplotlib.pyplot as plt
```

```python
path_scopus = r"C:\Users\Lenovo\Desktop\MetricSci\Scopus\Article.csv"
path_wos = r"C:\Users\Lenovo\Desktop\MetricSci\WoS\Article.csv"
articulo_scopus = pd.read_csv(path_scopus, encoding='utf-8')
articulo_wos = pd.read_csv(path_wos, encoding='utf-8')
```

```python
wos_main_articles = articulo_wos[articulo_wos['ismainarticle'] == True]
unique_wos_articles_raw = set(wos_main_articles['title'].unique()) # Mantengo 'raw' para claridad

scopus_main_articles = articulo_scopus[articulo_scopus['ismainarticle'] == True]
unique_scopus_articles_raw = set(scopus_main_articles['title'].unique()) # Mantengo 'raw' para claridad

# Definir valores nulos a excluir (cadenas o np.nan)
null_values_to_exclude = {np.nan, 'nan', 'NaN', 'NONE', 'None', 'none', '', ' ', '-'}

# Limpiar los conjuntos de DOIs principales de nulos
unique_wos_articles_cleaned = {
    x for x in unique_wos_articles_raw
    if str(x).lower() not in null_values_to_exclude and x is not np.nan
}
unique_scopus_articles_cleaned = {
    x for x in unique_scopus_articles_raw
    if str(x).lower() not in null_values_to_exclude and x is not np.nan
}


wos_count = len(unique_wos_articles_cleaned)
scopus_count = len(unique_scopus_articles_cleaned)


common_types = unique_wos_articles_cleaned & unique_scopus_articles_cleaned
num_common_types = len(common_types) 


all_unique_main_articles_union = unique_wos_articles_cleaned | unique_scopus_articles_cleaned
total_unique_main_articles = len(all_unique_main_articles_union)

table_1 = pd.DataFrame({
    "WoS": [wos_count],
    "Scopus": [scopus_count],
    "Total": [total_unique_main_articles], # Este es el total de la unión
})

table_1
```

```python
null_document_types_to_exclude = {np.nan, 'nan', 'NaN', 'N/A', 'n/a', 'none', 'None', '', ' ', '-'}

# --- Process WoS ---
wos_dt_exploded = articulo_wos[['doi', 'document_type']].assign(
    DT=articulo_wos["document_type"].astype(str).str.split(";")
).explode("DT")

# Corrected cleaning for WoS: use .isin() to create a boolean Series
wos_dt_cleaned = wos_dt_exploded[
    ~wos_dt_exploded['DT'].astype(str).str.lower().isin(null_document_types_to_exclude)
].dropna(subset=['DT'])

# --- Process Scopus ---
scopus_dt_exploded = articulo_scopus[['doi', 'document_type']].assign(
    DT=articulo_scopus["document_type"].astype(str).str.split(";")
).explode("DT")

# Corrected cleaning for Scopus: use .isin() to create a boolean Series
scopus_dt_cleaned = scopus_dt_exploded[
    ~scopus_dt_exploded['DT'].astype(str).str.lower().isin(null_document_types_to_exclude)
].dropna(subset=['DT'])

# --- Combine and Count Unique Articles per Type ---
combined_dt_df = pd.concat([wos_dt_cleaned, scopus_dt_cleaned], ignore_index=True)

# Drop duplicates based on BOTH 'doi' and 'DT'
unique_article_dt_combinations = combined_dt_df.drop_duplicates(subset=['doi', 'DT'])

# Count how many unique articles fall into each document type
document_counts = unique_article_dt_combinations["DT"].value_counts().reset_index()
document_counts.columns = ["Article Type", "Total"]

# Calculate percentage based on the new total of unique article-type associations
total_unique_combinations = document_counts["Total"].sum()
document_counts["Percentage"] = round((document_counts["Total"] / total_unique_combinations) * 100, 2)

document_counts = document_counts.sort_values(by="Total", ascending=False).reset_index(drop=True)


document_counts
```

```python
years_range = [2003, 2025]
years_full = np.arange(years_range[0], years_range[1] + 1)

articulo_wos["year"] = pd.to_numeric(articulo_wos["year"], errors='coerce').fillna(0).astype(float)
articulo_scopus["year"] = pd.to_numeric(articulo_scopus["year"], errors='coerce').fillna(0).astype(float) 

articulo_wos["total_times_cited"] = pd.to_numeric(articulo_wos["total_times_cited"], errors='coerce').fillna(0).astype(int)
articulo_scopus["cited_by"] = pd.to_numeric(articulo_scopus["cited_by"], errors='coerce').fillna(0).astype(int)

figure_1_data = pd.DataFrame()


for year in years_full:
   
    filtered_wos_articles = articulo_wos[(articulo_wos["year"] == year) & (articulo_wos["ismainarticle"] == True)]
    filtered_scopus_articles = articulo_scopus[(articulo_scopus["year"] == year) & (articulo_scopus["ismainarticle"] == True)]

    total_wos = len(filtered_wos_articles)
    total_scopus = len(filtered_scopus_articles)

    titles_wos = filtered_wos_articles["title"]
    titles_scopus = filtered_scopus_articles["title"]

    combined_titles = pd.concat([titles_wos, titles_scopus])

    unique_combined_titles = combined_titles.drop_duplicates()

    total_wos_scopus_unique = len(unique_combined_titles)

    new_row = pd.DataFrame({
        "PY": [year],
        "total": [total_wos_scopus_unique],
        "scopus": [total_scopus],
        "wos": [total_wos]
    })
    figure_1_data = pd.concat([figure_1_data, new_row], ignore_index=True)

figure_1_data = figure_1_data.sort_values("PY", ascending=False)

TC_all = pd.DataFrame(columns=["PY", "TC_sum_all", "TC_percentage"])

for year in years_full:
    ncitas_wos = articulo_wos[(articulo_wos["year"] == year) & (articulo_wos["ismainarticle"] == True)]["total_times_cited"].sum()
    ncitas_scopus = articulo_scopus[(articulo_scopus["year"] == year) & (articulo_scopus["ismainarticle"] == True)]["cited_by"].sum()
    ncitas = ncitas_wos + ncitas_scopus
    new_row_tc = pd.DataFrame({"PY": [year], "TC_sum_all": [ncitas], "TC_percentage": [None]})
    TC_all = pd.concat([TC_all, new_row_tc], ignore_index=True)

total_citas = TC_all["TC_sum_all"].sum()
TC_all["TC_percentage"] = round((TC_all["TC_sum_all"] / total_citas) * 100, 2)


TC_all["TC_sum_all"] = TC_all["TC_sum_all"].astype(int)

TC_all = TC_all.sort_values("PY", ascending=False)

print("Preprocesamiento completado.")
```

```python
# Extraer valores para la gráfica
tcx = TC_all["PY"].values
tcy = TC_all["TC_sum_all"].astype(int).values

tpx = figure_1_data["PY"].values
tpy = figure_1_data["total"].astype(int).values

wx = figure_1_data["PY"].values
wx1 = wx - 0.3  # Desplazamiento de barras WoS
wy = figure_1_data["wos"].astype(int).values

sx = figure_1_data["PY"].values
sx1 = sx + 0.5  # Desplazamiento de barras Scopus
sy = figure_1_data["scopus"].astype(int).values

print("Datos listos para la visualización.")
```

```python
legend_elements = [
    Line2D([0], [0], color='red', marker='o', label='Total Publications'),
    Line2D([0], [0], color='purple', marker='o', label='Total Citations'),
    Patch(facecolor='green', alpha=0.5, label='Scopus'),
    Patch(facecolor='orange', alpha=0.8, label='WoS')
]

# Filtrar los datos para incluir solo años
start_year = years_range[0]
mask_years = tpx >= start_year

# Aplicar la máscara a todas las variables correspondientes
tpx, tpy = tpx[mask_years], tpy[mask_years]
tcx, tcy = tcx[mask_years], tcy[mask_years]
wx1, wy = wx1[mask_years], wy[mask_years]
sx, sy = sx[mask_years], sy[mask_years]

# Configuración del gráfico
fig, ax = plt.subplots(figsize=(15, 7.5))

# Línea roja para total de publicaciones
ax.plot(tpx, tpy, color='r', marker='o', label='Total Publications')

# Etiquetas de ejes
ax.set_xlabel('Year', fontsize=18)
ax.set_ylabel('Total Publications', color='r', fontsize=18)

# Barras de WoS y Scopus
barw = 0.5
ax.bar(sx, sy, color='g', label='Scopus', alpha=0.5, width=barw)
ax.bar(wx1, wy, color='orange', label='WoS', alpha=0.8, width=barw)

# Segundo eje para citas totales
twin_axes = ax.twinx()
twin_axes.plot(tcx, tcy, color='purple', marker='o', label='Total Citations')
twin_axes.set_ylabel('Total Citations', color='purple', fontsize=20)
ax.set_ylim(0, max(tpy) * 1.1)
twin_axes.set_ylim(0, max(tcy) * 1.1)

# Título y leyendas
plt.title('Total Scientific Production vs. Total Citations', fontsize=23)
plt.legend(handles=legend_elements, loc='upper left', fontsize=14, frameon=False)

# Ajustar etiquetas del eje X
ax.set_xticks(tpx)
ax.set_xticklabels(tpx.astype(int), fontsize=14, rotation=75)

# Aumentar tamaño de los valores en todos los ejes
ax.tick_params(axis='x', labelsize=14)         # Eje X
ax.tick_params(axis='y', labelsize=14)         # Eje Y izquierdo
twin_axes.tick_params(axis='y', labelsize=14)  # Eje Y derecho

# Título y leyendas con tamaño de fuente aumentado
plt.title('Total Scientific Production vs. Total Citations', fontsize=23)

# Etiquetas en las barras y puntos
for i, label in enumerate(tcy):
    twin_axes.annotate(label, (tcx[i], tcy[i] + 0.5), color='purple', size=12)
for i, label in enumerate(tpy):
    ax.annotate(label, (tpx[i], tpy[i] + 1), xytext=(-10, 0), textcoords='offset points', color='red', size=12)
for i, label in enumerate(wy):
    ax.annotate(label, (wx1[i], wy[i] + 0.1), color='brown', size=12)
for i, label in enumerate(sy):
    ax.annotate(label, (sx[i], sy[i] + 0.2), color='green', size=12)

# Guardar figura
plt.savefig("Produccion_Cientifica/figura_1_1.svg")
plt.show()
```
